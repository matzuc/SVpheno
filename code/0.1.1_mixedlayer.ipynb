test for 1 year

```python

year = 1998


for year in range(1998, 2023):
  cm.subset(
    dataset_id="cmems_mod_arc_phy_my_topaz4_P1D-m",
    force_dataset_version="202211",
    variables=["mlotst"],
    username=user,
    password=pss,
    minimum_longitude=xmin,
    maximum_longitude=xmax,
    minimum_latitude=ymin,
    maximum_latitude=ymax,
    start_datetime=str(year)+"-01-01T00:00:00",
    end_datetime=str(year)+"-12-31T00:00:00",
    output_filename = "cmems_mod_arc_phy_anfc_6km_detided_P1D-m_mlotst_" +str(year) +".nc",
    output_directory = outdir,
    minimum_depth=0,
    maximum_depth=0,
    force_download = True
  )
  print("year " + str(year) + " done")  
  
```

```python
# read all data
ds = xr.open_mfdataset(path + '/mixdepth/*.nc', combine='by_coords')
ds
```

```python
# yearly average
ds_y = ds.groupby('time.year').mean('time')
ds_y

# plot first 4 years
ds_y.mlotst.sel(year=slice(1998, 2001)).plot(col='year', col_wrap=2)

# save yearly average
ds_y.to_netcdf(path + '/0.1.1_mixdepth_yearly_average.nc')
```

```python
# 10th quantile per year
ds_y10 = ds.groupby('time.year').quantile(0.1, dim='time')
ds_y10

# plot first 4 years
ds_y10.mlotst.sel(year=slice(1998, 2001)).plot(col='year', col_wrap=2)
ds_y10.write_netcdf(path + '/0.1.2_mixdepth_yearly_10th_quantile.nc')
```

```python
# 10th quantile per year
ds_y90 = ds.groupby('time.year').quantile(0.9, dim='time')
ds_y90

# plot first 4 years
ds_y90.mlotst.sel(year=slice(1998, 2001)).plot(col='year', col_wrap=2)
#ds_y90.write_netcdf(path + '/0.1.2_mixdepth_yearly_90th_quantile.nc')
```
